# 3. Create appropriate graphs to visually represent the relationship between financial services accessed (non, momible, both) and age, gender, marital status, land ownership and type of income. 

```python
sns.catplot(x ="mobile_money_classification", y="Q1", data=mobile_data);
plt.xlabel("Mobile money classification")
plt.ylabel("Age")
plt.title('Mobile money classification vs Age')
```

```python
sns.boxplot('mobile_money_classification', 'Q1', data=mobile_data)
plt.xlabel("Mobile money classification")
plt.ylabel("Age")
plt.title('Mobile money classification vs Age')
```

```python
sns.catplot('mobile_money_classification', 'Q2', data=mobile_data)
plt.xlabel("Mobile money classification")
plt.ylabel("Gender")
plt.title('Mobile money classification vs Gender')
```

```python
sns.lmplot('mobile_money_classification', 'Q3', data=mobile_data)
plt.xlabel("Mobile money classification")
plt.ylabel("Marital Status")
plt.title('Mobile money classification vs Marital Status')
```

```python
sns.lmplot('mobile_money_classification', 'Q8_1', data=mobile_data)
plt.xlabel("Mobile money classification")
plt.ylabel("Salaries/wages")
plt.title('Mobile money classification vs Salaries/wages')
```
